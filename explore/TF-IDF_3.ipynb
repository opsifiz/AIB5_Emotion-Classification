```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
# print(stopwords.words('english'))
```

```python
df = pd.read_csv('drive/MyDrive/train_dataset2.csv')
df
```

```python
df.describe()
```

```python
grouped = df.groupby('status')['text'].apply(lambda texts: ' '.join(texts))
```

```python
grouped.head(10)
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    words = text.split()  # Tokenize and lowercase
    filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
    return ' '.join(filtered_words)
```

```python
all_status = grouped.index.to_list()
all_status
```

```python
for status in all_status:
  grouped[status] = remove_stopwords(grouped[status])
```

```python
grouped.head(10)
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(grouped)
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(),
                        index=grouped.index,
                        columns=vectorizer.get_feature_names_out())
```

```python
top_n = 10
for status in tfidf_df.index:
    print(f"\nTop words for status '{status}':")
    top_words = tfidf_df.loc[status].sort_values(ascending=False).head(top_n)
    print(top_words)
```

```python
top_n = 10
for status in tfidf_df.index:
    print(f"\nTop words for status '{status}':")
    top_words = tfidf_df.loc[status].sort_values(ascending=True).head(top_n)
    print(top_words)
```
